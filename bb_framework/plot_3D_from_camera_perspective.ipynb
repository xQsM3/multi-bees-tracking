```python
import plotly
import pandas as pd
import plotly.express as px
import plotly.io as pio
pio.renderers.default = 'browser'
import numpy as np
```

```python
direc1 = "/media/linx123-rtx/Elements/test5/day1/analysis/3Dtracks/V13 to foraging (single)_CR3000x2 1838-ST-159160_1.csv"
direc2 = "/media/linx123-rtx/Elements/test5/day1/analysis/3Dtracks/V13 to foraging (single)_CR3000x2 1838-ST-159161_1.csv"
transform = False


R = np.array([[-0.01569649,  0.99946256,  0.02877858],
 [-0.12604994, -0.03053046,  0.99155398],
 [ 0.99189971,  0.01193638,  0.12646142]])

T = np.array([[4.44749048],
 [-1793.30879377],
 [1603.7252569 ]])

K = np.array([[3.00e+03, 0.00e+00, 8.48e+02],
 [0.00e+00, 3.00e+03, 8.55e+02],
 [0.00e+00, 0.00e+00, 1.00e+00]])


df = pd.read_csv(direc1)
df["size"] = [2]*len(df["x"])
df["camerapair"] = [1]*len(df["x"])
df2 = pd.read_csv(direc2)
df2["size"] = [2]*len(df2["x"])
df2["camerapair"] = [2]*len(df2["x"])



if transform == False:
    Tmatrix = np.append(np.append(R,T,axis=1),np.array([0,0,0,1])).reshape(4,4)
    for i,(x,y,z) in enumerate(zip(df["x"],df["y"],df["z"])):
        pt = np.array([x,y,z,1])
        # transformation
        Tpt = Tmatrix.dot(pt)
        Tpt = Tpt / Tpt[3]
        #print(Tpt)
        df.at[i,'x']=Tpt[0]
        df.at[i,'y']=Tpt[1]
        df.at[i,'z']=Tpt[2]
    for i,(x,y,z) in enumerate(zip(df2["x"],df2["y"],df2["z"])):
        pt = np.array([x,y,z,1])
        #print(pt)
        # transformation
        Tpt = Tmatrix.dot(pt)
        print(Tpt)
        Tpt = Tpt / Tpt[3]
        #print(Tpt)
        df2.at[i,'x']=Tpt[0]
        df2.at[i,'y']=Tpt[1]
        df2.at[i,'z']=Tpt[2]        
        
    df = df.append(df2)
    df["ID"] = df["ID"].astype(str)
    # plot 3d plot
    fig = px.scatter_3d(df, x='x', y='y', z='z',
              color='camerapair', size='size', size_max=8,
              symbol='ID', opacity=0.7)
    
    fig.update_layout(
        scene = {"aspectmode":"manual","aspectratio":dict(y=1, x=8/3, z=1),
                "yaxis":dict(nticks=4, range=[-100,200],),
                "xaxis":dict(nticks=4, range=[500,-300],),
                "zaxis":dict(nticks=4, range=[1960,1660],)})
    
    fig.update_layout()

    camera = dict(
        up=dict(y=0, x=0, z=0),
        center=dict(y=0, x=0, z=-21), # z is the focal length in millimeter, f_mm = f_pixel * sensor_height_mm / image_height_pixel
        eye=dict(x=0, y=0, z=0)
    )

    fig.update_layout(scene_camera=camera)
    fig.show()
else:
    df = df.append(df2)
    df["ID"] = df["ID"].astype(str)
    # plot 3d plot
    fig = px.scatter_3d(df, x='x', y='y', z='z',
              color='camerapair', size='size', size_max=8,
              symbol='ID', opacity=0.7)

    fig.update_layout(
        scene = {"aspectmode":"manual","aspectratio":dict(x=1, y=8/3, z=1),
                "xaxis":dict(nticks=4, range=[30,-270],),
                "yaxis":dict(nticks=4, range=[-400,400],),
                "zaxis":dict(nticks=4, range=[1960,1660],)})
    fig.update_layout()

    camera = dict(
        up=dict(y=0, x=0, z=0),
        center=dict(y=0, x=0, z=-21),# z is the focal length in millimeter, f_mm = f_pixel * sensor_height_mm / image_height_pixel
        eye=dict(x=0, y=0, z=0)
    )

    fig.update_layout(scene_camera=camera)
    fig.show()
```
